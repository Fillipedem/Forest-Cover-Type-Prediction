### Analise Exploratória

```python
import numpy as np
import pandas as pd
import seaborn as sns

### Plot
import matplotlib.pyplot as plt
### Parametros
plt.rcParams["figure.figsize"] = (20,10)
plt.rcParams.update({'font.size': 22})
```

### Read dataset

```python
df = pd.read_csv('./data/train.csv', index_col=0)
df.head()
```

```python
df.describe()
```

```python
# Checando por missing values
df.isnull().sum()
```

Sem missing data

### Distribuição das Classes

```python
dist = df['Cover_Type'].value_counts()
plt.bar(dist.index, dist)
```

### Attr -  Soil Type



```python
### checamos a quantidade de vezes que cada tipo de solo aparece
qtd_soil = df.loc[:, 'Soil_Type1':'Soil_Type40'].sum()

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar([i for i in range(1, 41)], qtd_soil.values)
plt.title("Quantidade de vezes que um Tipo de Solo aparece na base de dados")
plt.show()
```

   Considerando que existem poucos exemplos de muitos tipos de solo, o ideal será criar categorias mais amplas que diminuam o número e o tipo de solos.

### Wilderness Area

```python
for i in [1,2,3,4]:
    sns.barplot(x="Cover_Type", y="Wilderness_Area" + str(i), data=df)
    plt.show()
```

### Correlação

```python
# criando dataframe sem as variaveis de SoilType
soil_type_list = list(df.columns[-41:-1])
df_tp = df.drop(columns=soil_type_list)
```

```python
# plot heatmap correlação
corr = df_tp.corr()
sns.heatmap(corr, cmap=sns.diverging_palette(20, 220, n=200))
```
